```python
turkish_chars = {
'ğ':'g',
'ü':'u',
'ı':'i',
'ş':'s',
'ç':'c',
'ö':'o'}
```

```python
def replace_turkish_chars(text):
    for k,v in turkish_chars.items():
        text = text.replace(k, v)
    return text
```

```python
data_reviews['comment'] = data_reviews['comment'].str.lower()
data_reviews['comment_non_turk_chars'] = data_reviews['comment'].apply(replace_turkish_chars)
```

```python
def look_for_keywords(text, corpus):
    keywords = 0
    for word in corpus:
        if word in text:
            keywords += 1 
    return keywords
```

# preprocessing and vectorizing

```python
nltk.download('stopwords')
stop_word_list = nltk.corpus.stopwords.words('turkish')  
stop_word_list = [replace_turkish_chars(i) for i in stop_word_list]
```

```python
def remove_stopwords(tokens):
    filtered_tokens = [token for token in tokens if token not in stop_word_list]#stop word'lerden temizlenir.  
    return filtered_tokens
```

```python
def remove_word_contains_numeric(text):
    return re.sub(r'\w*\d\w*', ' ', text)
```

```python
def remove_nonword_chars(text):
    return re.sub(r'[\W_]+', ' ', text)
```

```python
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
```

```python
def remove_emojis(text):
    return emoji_pattern.sub(r'', text)
```
