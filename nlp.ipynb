```python
turkish_chars = {
'ğ':'g',
'ü':'u',
'ı':'i',
'ş':'s',
'ç':'c',
'ö':'o'}
```

```python
def replace_turkish_chars(text):
    for k,v in turkish_chars.items():
        text = text.replace(k, v)
    return text
```

```python
def look_for_positives(text):
    pos = 0
    for p in positive:
        if p in text:
            pos += 1 
    return pos
```

```python
def look_for_negatives(text):
    neg = 0
    for n in negative:
        if n in text:
            neg += 1   
    return neg
```

```python
nltk.download('stopwords')
stop_word_list = nltk.corpus.stopwords.words('turkish')  
stop_word_list = [replace_turkish_chars(i) for i in stop_word_list]
```

```python
def remove_stopwords(tokens):
    filtered_tokens = [token for token in tokens if token not in stop_word_list]#stop word'lerden temizlenir.  
    return filtered_tokens
```

```python
def remove_word_contains_numeric(text):
    return re.sub(r'\w*\d\w*', ' ', text)
```

```python
def remove_nonword_chars(text):
    return re.sub(r'[\W_]+', ' ', text)
```

```python
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
```

```python
def remove_emojis(text):
    return emoji_pattern.sub(r'', text)
```

```python
def lemmatize_verbs(words):
    lemmatizer = zeyrek.MorphAnalyzer()
    lemmas = []
    for word in words:
        lemma = lemmatizer.lemmatize(word)[0][0]
        lemmas.append(lemma)
    return lemmas
```

```python
def preprocess(text):
 
    text = text.lower()
    text = remove_word_contains_numeric(text)
    text = remove_emojis(text)
    text = remove_nonword_chars(text)
    text = text.strip()
    text = text.split(' ')
    text = remove_stopwords(text)
    text = lemmatize_verbs(text)
    text = [word for word in text if len(word) < 25 and len(word) >= 2]
    text = ' '.join(text)
    
    return text
```

```python
data_reviews['preprocessed_text'] = data_reviews['comment'].apply(preprocess)
```
